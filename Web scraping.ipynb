```python
url=['https://www.cars24.com/buy-used-cars-hyderabad/?itm_source=Cars24Website&itm_medium=sticky_header',
 'https://www.cars24.com/buy-used-car?sort=P&storeCityId=2130&pinId=302001',
'https://www.cars24.com/buy-used-car?sort=P&storeCityId=5732&pinId=600001',
'https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId=400001&pinId=400001']
```

```python
for i in range(2, 4):
    url.append('https://www.cars24.com/buy-used-cars-hyderabad/?itm_source=Cars24Website&itm_medium=sticky_header&page={}'. format(i))
for i in range(2,15):
    url.append('https://www.cars24.com/buy-used-car?sort=P&storeCityId=2130&pinId=302001&page={}'.format(i))
for i in range(2,15):
    url.append('https://www.cars24.com/buy-used-car?sort=P&storeCityId=5732&pinId=600001&page={}'.format(i))
for i in range(2,23):
    url.append('https://www.cars24.com/buy-used-car?sort=P&storeCityId=2378&pinId=400001&page={}'.format(i))
    
    
```

```python
carname1 = []
Type1 = []
price1 = []
emi1 = []
details1= []
discount1 = []
location1 = []

for i in url:
#URL = 'https://www.cars24.com/buy-used-cars-hyderabad/?itm_source=Cars24Website&itm_medium=sticky_header&page1={}'. format(i)
    
    page = requests.get(i)
    htmlCode = page.text
    
    soup = BeautifulSoup(htmlCode)
    
    for x in soup.find_all('div', attrs={'class' : 'col-4'}):

        carname = x.find('h2', attrs={'class' : '_3FpCg'})
        if carname is None:
            carname1.append(np.NaN)
        else:
            carname1.append(carname.text)

        Type = x.find('p', attrs={'class' : 'cvakB'})
        if Type is None:
            Type1.append(np.NaN)
        else:
            Type1.append(Type.text)

        price = x.find('div', attrs={'class' : '_7udZZ'})
        if price is None:
            price1.append(np.NaN)
        else:
            price1.append(price.text)
            
        emi = x.find('div', attrs={'class' : '_2HFRN'})
        if emi is None:
            emi1.append(np.NaN)
        else:
            emi1.append(emi.text)
            
        details = x.find('ul', attrs={'class' : 'bVR0c'})
        if details is None:
            details1.append(np.NaN)
        else:
            details1.append(details.text)
        
        discount = x.find('div', attrs={'class' : '_3mZ8j'})
        if discount is None:
            discount1.append(np.NaN)
        else:
            discount1.append(discount.text)
            
        if 'hyderabad' in i:
            location1.append('Hyderabad')
        elif 'CityId=2130' in i:
            location1.append('Jaipur')
        elif 'CityId=5732' in i:
            location1.append('Chennai')
        elif 'CityId=2378' in i:
            location1.append('Mumbai')
        else: 
            location1.append(np.NaN)
            
 
```

```python
print(len(carname1))
print(len(Type1))
print(len(price1))
print(len(emi1))
print(len(details1))
print(len(discount1))
print(len(location1))
```

## USING REGEX TO MODIFY THE DATA

```python
import re
```

```python
df['Name'] = df['Name'].fillna(0)
```

```python
df['Name'].value_counts()
```

```python
regex = r'^2\d+'

df['Year'] = df['Name'].apply(lambda x : re.findall(regex, x) if x!='NaN' else x)
df.Year = df.Year.apply(lambda x: "".join(x))
```

```python
df['Type'] = df['Type'].fillna(0)
```

```python
df.Type
```

```python
regex = r'(?:Manual|Automatic)'

df['cartype'] = df['Type'].apply(lambda x : re.findall(regex, x) if x!='NaN' else x)
df.cartype = df.cartype.apply(lambda x: "".join(x))
```

```python
c=[]
for i in df['Name'] :
    if i == 0:
        c.append(0)
    else:
        c.append(i.split()[2:])
```

```python
df['Model']=c
df.Model = df.Model.apply(lambda x: " ".join(x) if x!=0 else "None")
```

```python
df.Model.value_counts()
```

```python
regex = r'(?:Hyundai|Renault|Maruti)'

df['Brand'] = df['Name'].apply(lambda x : re.findall(regex, x) if x!= np.NaN else x)
df.Brand = df.Brand.apply(lambda x: "".join(x))
```

```python
df.Brand.value_counts()
```

```python
df.Brand
```

```python
df['Type'].replace(to_replace=["Manual","Automatic"],value="",inplace=True,regex=True)
```

```python
df.Type
```

```python
df.Details.value_counts()
```

```python
df['Details'] = df['Details'].fillna(0)
```

```python
regex = r'([,\d]+\skm)'
df['km'] = df.Details.apply(lambda x : re.findall(regex,x) if x!=0 else "None")
df.km = df.km.apply(lambda x: "".join(x))
```

```python
df['km'].replace(to_replace="km",value="",inplace=True,regex=True)
```

```python
df.km.value_counts()
```

```python
regex = r'(?:1st|2nd)'
df['owner'] = df.Details.apply(lambda x : re.findall(regex,x) if x!=0 else "None")
df.owner = df.owner.apply(lambda x: "".join(x))
```

```python
df['owner'].replace(to_replace=["st","nd"],value="",inplace=True,regex=True)
```

```python
df.owner.value_counts()
```

```python
regex = r'(?:Petrol|Diesel)'
df['fuel'] = df.Details.apply(lambda x : re.findall(regex,x) if x!=0 else "None")
df.fuel = df.fuel.apply(lambda x: "".join(x))
```

```python
df.Price
```

```python
df['Price'] = df['Price'].fillna(0)
```

```python
df['Price'].replace(to_replace=["₹",","],value="",inplace=True,regex=True)
```

```python
df.Price=df.Price.astype(int)
```

```python
df['Discount'] = df['Discount'].fillna(0)
```

```python
df['Discount'].replace(to_replace=["₹",",","OFF"],value="",inplace=True,regex=True)
```

```python
df['EMI'] = df['EMI'].fillna(0)
```

```python
df['EMI'].replace(to_replace=["₹",",","/","month"],value="",inplace=True,regex=True)
```

```python
df.drop(["Name","Details"],axis=1,inplace=True)
```
